```python
#5.

import tkinter as tk

form = tk.Tk()
form.title("5. Ders")

def topla():
    label = tk.Label(form,text= "The weird story of Benjamin Button",fg="orange",font="Times 12 bold")
    label.pack(side=tk.LEFT)

def yeni_pencere():
    root = tk.Tk()
    root.title("Yeni Pencere")
    root.geometry("500x400")
    root.state("zoomed")
    label = tk.Label(root,text="bu yeni bir pencere",font="calibri 15 italic")
    # eğer root yazmasaydık ilk sayfada bu yeni bir pencere yazacaktı
    label.pack()
    
    
form.geometry("500x400")
buton = tk.Button(form,text= "Benjamin Button",fg="black",bg="red",font="Times 15 italic",command = topla)
buton.pack(side="top")

buton2 = tk.Button(form,text="Another button",fg="white",bg ="black",font="Times 15 bold",command=yeni_pencere)
buton2.pack(side="top")
form.mainloop()
```

```python
#7.
import tkinter as tk

form = tk.Tk()
form.title("7 Entry")
form.geometry("500x450")





def big():
    form.state("zoomed")

def close():
    form.destroy()

def takedw():
    form.state("iconic")
    
    
closer = tk.Button(form,text="X",fg="red",bg="black",command=close)
closer.pack(side ="right",anchor="ne")
    
bigger = tk.Button(form,text = "▢",fg="white",bg="black",command = big)
bigger.pack(side ="right",anchor="ne")

take_down = tk.Button(form,text="_",fg="white",bg="black",command= takedw)
take_down.pack(side="right",anchor="ne")


label2 = tk.Label(form,text="Birinci sayıyı giriniz:",fg="black",font="Times 12 bold")
label2.pack(side="top",fill =tk.X,pady=30)
entry = tk.Entry(form,fg="black",bg="brown",font="Times 12 italic")
entry.pack(side = "top",fill=tk.X)

label3 = tk.Label(form,text="İkinci sayıyı giriniz:",fg="black",font="Times 12 bold")
label3.pack(side="top",fill =tk.X)
entry2 = tk.Entry(form,fg="black",bg="red",font="Times 12 italic")
entry2.pack(side ="top",fill =tk.X)


def summer():
    x = entry.get()
    y = entry2.get()
    
    summing = x+y
    summing_int = int(x)+int(y)
    label = tk.Label(form,text = "summing as string= {}".format(summing),font="Times 11")
    label.pack()
    label1 = tk.Label(form,text = "summing as integer= {}".format(summing_int),font="Times 11")
    label1.pack()

    
def nothing():
    
    label4 = tk.Label(form,text="There is no function in this button")
    label4.pack()
button = tk.Button(form,text="Sum",bg="black",fg="yellow",font = "Times 12 bold",command = summer)
button.pack(side ="bottom",fill =tk.X)
button1 = tk.Button(form,text= "Benjamin Button",fg="red",bg="black",font="Times 12 bold",command = nothing)
button1.pack(expand = 1,anchor="center",ipadx =100,ipady=8)
# paddy = away from distriction
# ipadx = ,ipady size




# anchor n up,s down, w left, ne north east,e east, center in the center
#expand 0-1

#x ekseni boyunca kaplar bottom en altta olması için

form.mainloop()
```
